# Toggle – operador de realce

O operador de realce Toggle é uma técnica de processamento de imagem utilizada para melhorar o contraste de uma imagem. Ele funciona aplicando uma operação de realce que destaca as diferenças de intensidade entre os pixels da imagem, tornando as características mais visíveis. Esse operador é especialmente útil em imagens de baixo contraste, onde os detalhes podem ser difíceis de distinguir. No exemplo abaixo, é aplicado o operador de realce Toggle a uma imagem de pólen com baixo contraste, utilizando um kernel de tamanho (50, 50).

```python
image = read_image("angiogr.tif")
processing_image = processings.toggle_highlight(image, kernel_size=(40, 40))

plot_images(
    titles=["Original", "Toggle Highlight"],
    images=[image, processing_image],
)
```

# Filtro Alternado Sequencial

O Filtro Alternado Sequencial é uma técnica de processamento de imagem que combina operações de abertura e fechamento para remover ruídos e suavizar a imagem. Este filtro é particularmente útil para melhorar a qualidade de imagens que contêm ruídos ou pequenas imperfeições. No exemplo abaixo, o filtro alternado sequencial é aplicado a uma imagem utilizando uma sequência de operações de abertura e fechamento com um número de iterações definido.

Abaixo, a imagem original e a imagem processada utilizando o filtro alternado sequencial são exibidas lado a lado para comparação.

```python
image = read_image("mascara.tif")
processing_image = processings.alternating_sequential_filter(image, seq='OC', n=10)

plot_images(
    titles=["Original", "Alternating Sequential Filter"],
    images=[image, processing_image],
)
```
