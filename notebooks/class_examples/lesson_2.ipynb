# Gradiente morfológico

O gradiente morfológico é uma operação de processamento de imagem que realça os contornos dos objetos presentes na imagem. Ele é obtido pela diferença entre a dilatação e a erosão da imagem original. Essa técnica é amplamente utilizada em segmentação e análise de imagens, especialmente em aplicações biomédicas, onde a detecção precisa dos contornos das células é crucial para diagnósticos e estudos.

No exemplo abaixo, o gradiente morfológico é aplicado a uma imagem de células sanguíneas, utilizando um kernel de tamanho (3, 3). A imagem resultante destaca os bordos das células, facilitando a visualização e análise dos elementos presentes na imagem.

```python
image = read_image("bloodcells.tif")
processing_image = processings.morphological_gradient(image, kernel_size=(3, 3))

plot_images(
    titles=["Original", "Morphological Gradient"],
    images=[image, processing_image],
)
```
