# Dilatação

A dilatação é uma operação morfológica em processamento de imagens que aumenta as áreas claras de uma imagem binária. Ela é realizada aplicando um elemento estruturante (geralmente uma forma simples, como um quadrado ou círculo) sobre a imagem. Para cada pixel da imagem, a operação verifica se a forma do elemento estruturante se ajusta ao redor dele, e se algum pixel da forma sobrepõe uma área branca, o pixel central é alterado para branco. Isso faz com que as regiões brancas se expandam, "dilatando" as formas presentes na imagem.

## Dilatação 1D horizontal

```python
image = read_image("astablet.tif")
processing_image = processings.dilation(image, kernel_size=(1, 3))

plot_images(
    titles=["Original", "Dilation"],
    images=[image, processing_image],
)
```

## Dilatação 2D

```python
image = read_image("pcb_gray.tif")
processing_image = processings.dilation_elipse(image, kernel_size=(10, 10))

plot_images(
    titles=["Original", "Dilation"],
    images=[image, processing_image],
)
```

# Erosão

A erosão é uma operação morfológica em processamento de imagens que diminui as áreas claras de uma imagem binária. Ela é realizada aplicando um elemento estruturante sobre a imagem. Para cada pixel da imagem, a operação verifica se a forma do elemento estruturante se ajusta ao redor dele, e se todos os pixels da forma sobrepõem uma área branca, o pixel central é alterado para preto. Isso faz com que as regiões brancas se contraiam, "erodindo" as formas presentes na imagem.

```python
image = read_image("pcb_gray.tif")
processing_image = processings.erosion_elipse(image, kernel_size=(6, 6))

plot_images(
    titles=["Original", "Erosion"],
    images=[image, processing_image],
)
```
