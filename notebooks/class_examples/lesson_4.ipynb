# Remoção de pequenas regiões

Este exemplo apresenta a remoção de pequenas regiões em uma imagem binária. A imagem original é processada para remover regiões menores que um tamanho mínimo especificado. A imagem original e a imagem processada são exibidas lado a lado para comparação.

```python
image = read_image("circuit_bw.tif")
processing_image = processings.remove_small_regions(image, min_size=1000)

plot_images(
    titles=["Original", "Remove small regions"],
    images=[image, processing_image],
)
```

# Reconstrução

A reconstrução por dilatação é uma técnica utilizada em processamento de imagens para restaurar a forma original de uma região a partir de uma imagem máscara. Este processo é útil para remover ruídos e pequenas imperfeições, preservando as características principais da imagem.

No exemplo abaixo, a imagem original é processada utilizando uma máscara gerada por erosão. A imagem reconstruída é então comparada com a imagem original e a máscara para ilustrar o efeito da reconstrução por dilatação.

```python
image = read_image("text_128.tif")
marker = processings.erosion(image, kernel_size=(6, 2))

preprocessed_image = processings.reconstruction_by_dilation(image, marker)

plot_images(
    titles=["Original image", "Marker", "Reconstruction by dilation"],
    images=[image, marker, preprocessed_image],
)
```
