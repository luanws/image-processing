# Transformada H-máxima

A transformada h-máxima é uma técnica utilizada em processamento de imagens para destacar regiões de interesse em uma imagem, eliminando pequenos máximos locais que são menores que um valor h especificado. Essa técnica é útil para simplificar a imagem e facilitar a detecção de características importantes.
A transformada h-máxima de uma imagem `I` é definida como a supressão de todos os máximos locais cuja altura é menor que um valor `h`. Em outras palavras, ela reduz a intensidade dos picos que são menores que `h`, mantendo apenas os picos mais significativos.

```python
image = read_image("r4x2_256.tif")

preprocessed_image = processings.h_maxima_transform(image, 50)

plot_images(
    titles=["Original image", "H-maxima transform"],
    images=[image, preprocessed_image],
)
```

# Transformada H-mínima

A transformada H-minima é uma técnica utilizada no processamento de imagens para suprimir mínimos locais que são menos profundos do que um determinado valor de limiar, chamado de h. Em outras palavras, ela remove pequenas depressões na imagem que não são suficientemente profundas, preservando apenas os mínimos locais mais significativos.
Essa técnica é particularmente útil em tarefas de segmentação de imagens, onde é necessário identificar e separar objetos distintos. Ao aplicar a transformada H-minima, é possível reduzir o ruído e melhorar a precisão da segmentação, pois os mínimos locais irrelevantes são eliminados.
A transformada H-minima é frequentemente utilizada em conjunto com outras operações morfológicas, como a dilatação e a erosão, para refinar os resultados da segmentação e obter uma representação mais clara e precisa dos objetos presentes na imagem.

```python
image = read_image("r4x2_256.tif")

preprocessed_image = processings.h_minima_transform(image, 70)

plot_images(
    titles=["Original image", "H-minima transform"],
    images=[image, preprocessed_image],
)
```
